```python
df = pd.read_csv("Data/8_36_copy.csv", delimiter = ";")
```

```python
for i in range(df.shape[1]):
    mask = "Column"
    mask = mask + str(i + 1)
    mean = df[mask].mean()
    df[mask] = df[mask] - mean
```

```python
df.head(10)
```

```python
df_transp = df.T
```

```python
teta = (1/df_transp.shape[1]) * df.dot(df_transp)
```

```python
lambda1 = np.linalg.eig(teta)[0]
```

```python
lambda1
```

```python
lambda_sum = lambda1.sum()
```
